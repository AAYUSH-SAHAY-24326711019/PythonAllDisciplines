```python
#1 create the string
name = "Aayush"

#2 accessing the characters at the index
print("The 1st character is :",name[0])
```

```python
#3 codes of the case conversion
name = "aayush"
print("name.upper()",name.upper())
print("name.lower()",name.lower())
print("name.title()",name.title())
print("name.capitalize()",name.capitalize())
print("name.swapcase()",name.swapcase())
```

```python
#searching and checking the contents of the string
email = "admin123@example.com"
print("email.find('admin')",email.find("admin"))
print("email.index('d')",email.index("admin"))
print("email.startswith('admin')",email.startswith("admin"))
print("email.endswith('com')",email.endswith("com"))
```

```python
#4 checking the types of the strings given
print("'123'.isdigit()","123".isdigit())
print("'abc'.isalpha()","abc".isalpha())
print("'123abc'.isalnum()","123abc".isalnum())
print("'admin'.islower()","admin".islower())
print("'HELLO'.isupper()","HELLO".isupper())
```

```python
#5 modification of the strings
s="  python  "
print(r"The length of '  python  ' is",len(s))
print("s.strip()",s.strip(),"length=",len(s))
print("s.lstrip()",s.lstrip(),"length=",len(s))
print("s.rstrip()",s.rstrip(),"length=",len(s))
print("s.replace('python','java')",s.replace('python','java'))
```

```python
#6 slicing of the strings

strr = "123456"
print("strr[0:5]",strr[0:5]) # 0 is start end is 5-1=4
print('strr[:]',strr[:]) # from start to end
print('strr[:5]',strr[:5]) # from start to 4
print("reverse:",'strr[-1:]',strr[-1::-1]) #reverse
print("with step value 2",'strr[:4:2]',strr[:4:2])
```
