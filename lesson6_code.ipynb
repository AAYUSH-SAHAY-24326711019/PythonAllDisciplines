```python
'''
File modes

r-read w-write a-append(add at the end)
x- create a new file
b- binary mode file
t- text mode file
+- read and write in file

There are 2 functions open and close
that work with the files
'''
```

```python
#make a file and do create file x
import os
f=open(r"temp\file.txt","x")
```

```python
# write some data into file.txt
line_str = "This is the content inside file.txt"
f=open(r"temp\file.txt","w")
f.write(line_str)
```

```python
#Read the data from the file.txt
f=open(r"temp\file.txt","r")
print(f.read())
```

```python
# time to read line by line manner.
new_line_str1 = "\nThis is second line in file.txt"
new_line_str2 = "\nThis is third line file.txt"
f=open(r"temp\file.txt","a") # append mode
f.writelines(new_line_str1)
f.writelines(new_line_str2)
```

```python
# after writing , now trying to access the lines
f=open(r"temp\file.txt","r") 
print(f.readlines())
f.close() #returns a list to you
```

```python
#another way to write more files into it
lines = ["This is the 1st line\n",
         "This is the 2nd line\n",
         "This is the 3rd line\n"]

f=open(r"temp\file.txt","w")
f.writelines(lines)
f.close() #stores the lines line by line
```

```python
#use of the with keywoed

with open(r"temp\file.txt","r") as f:
    print(f.readlines())

f.close()
```

```python
#working with the binary files
with open(r"temp\image.jpg","rb") as f:
    print(f.readlines())

f.close()
```

```python
#tell the pointer location
with open(r"temp\file.txt","r") as f:
    print("Cursor currently at:",f.tell())
    #moving cursor at 10 units 
    f.seek(10)
    print("Cursor now currently at:",f.tell())

    print(f.readlines())

f.close()
```

```python
# to check whether the file exists

import os

print("status of file.txt :",os.path.exists(r"temp\file.txt"))
```

```python
#time to delete the image.jpg from the temp\
os.remove(r"temp\image.jpg")
```

```python
# time to rename file.txt to file_records_{date}.txt
from datetime import date
import os

today = date.today().isoformat()

file_name_new = f"temp\\file_records_{today}"
os.rename(r"temp\file.txt",file_name_new)
```

```python
#writing the code to read the data in the form of csv
import csv

with open(r"temp\data_records.csv","r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
#code to write into a json file

import json

data1 = {"name":"Aayush",
        "age":22,
        "location":"Patna"}

data2= {"name":"Gopal Ji",
        "age":22,
        "location":"Patna"}

with open(r"temp\data_temp.json","w") as f:
    json.dump(data1,f)
    json.dump(data2,f)
```
