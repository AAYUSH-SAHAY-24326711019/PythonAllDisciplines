python is a dynamically typed language, type decided in the runtime
then it is also a reason for why overloading is not possible

```python
demo = (1,2,3)
print("type of demo is :",type(demo))
```

```python
#making of the frozen sets
fs = frozenset([1,2,3,4])
print("type of fs is :",type(fs))
```

it means that once made then data cannot be added / altered
this is called the immutable types that are the sets, tuple, frozenset, strings, bytes

```python
# take the data and show its bytes value (used on files and networking)
binary_value = b"Byte Value is immutable"
print("The binary value is :",binary_value," type = ",type(binary_value))
```
