```python
#make the list from the range

d= list(range(5))
print(d)
```

```python
# print the list in the reversed manner
print(d[-1::-1])
```

```python
#modify the list by an index

d[0]=10
print(d)
```

```python
#ways to insert into the list
#append, insert and extend

d.append(20) #queue fashion
print(d)

# d.insert(index:2,value:10)
d.insert(2,10)
print(d)

d.extend([44,55]) #add list in queue fashion
print(d)
```

```python
#ways to remove the elements from the list
# clear, pop, pop(by index),remove(by value)

d.pop()
print(d) # pop will remove from the right hand last val

d.pop(0)
print(d) #pop in stack manner

d.remove(10)
print(d) #removes the value from list

d.clear()
print(d) #clear all contents of list
```

```python
#concatenate the list
a=[10,20]
b=[30,40]
print(a+b)
```

```python
#repeat the list
print(a*3)
```

```python
# iterate the list 2 ways, for each and enumerate
for val in (a*3):
    print(val,end=" ")
```

```python
#enumerate will iterate with index
for i,j in enumerate(a*3):
    print(f"({i},{j})",end=",") #i is for the inedex, j is for value
```

```python
# The built in functions in list

nums = range(10,21)
```

```python
print(len(nums))
print(max(nums))
print(min(nums))
print(sum(nums))
print(sorted(nums))
print(nums.count(10))
```

```python
# make the list of all square numbers
squares = [x*x for x in range(5)]
print(squares)
```

```python
#make the list of the even numbers
evens = [x for x in range(10) if x%2 ==0]
print(evens)
```

```python
#make the stack and the stack operations

stack= []
stack.append(10)
stack.append(20)
stack.append(30)
#display the contents
print(stack)

#tos is -1

stack.pop()
stack.pop()

#pop is from the -1
```

```python
#make the queue from the list

queue = []

queue.append(10)
queue.append(20)
queue.append(30) #add elements from the rear

#display the queue
print(queue)

queue.pop(0)
queue.pop(0) # remove the elements from the front

print(queue)
```

```python
# make the doubly ended queue

from collections import deque

l = list(range(1,10,2))

dq = deque(l)

print(dq) #display

dq.appendleft(0)
dq.append(11)

print(dq)

dq.popleft() # will pop from left
dq.pop() #will remove from the right

print(dq) #display
```

```python
#remove the list with duplicates
list_with_duplicates = [10,101,10,102,10,103]
new_list = list_with_duplicates.copy()
new_list=list(set(new_list))
print(new_list)
```
