```python
# Demonstration of the Inheritance as well as the
# access specifiers

#access specifiers
#no underscore public overall accessible
# _ (single underscore is the protected) at starting
# protected is inheritance accessible

class Parent:
    def __init__(self):
       self._val=100
 
#make the child class
class Child(Parent):
    def show(self):
        print("protected value is ",self._val)
   
c1= Child()
c1.show()
        
```

```python
#Types of inheritance

#single inheritance

class Parent:
    pass

class Child(Parent):
    pass
```

```python
#multilevel inheritance

class GrandParent:
    pass

class Parent(GrandParent):
    pass

class Child(Parent):
    pass
```

```python
#multiple inheritance

class Father:
    def IQ_from_Father(self):
        print("iq method from father")
class Mother:
    def Beauty_from_Mother(self):
        print("beauty method from mother")

class Child(Father,Mother):
    def Child_Details(self):
        self.IQ_from_Father()
        self.Beauty_from_Mother()

Alice_child = Child()
Alice_child.Child_Details()
```

```python
#hirearchical inheritance

class A:
    pass

class B(A):
    pass

class C(A):
    pass

# A is super class and B and C inherit from there
```

```python
# method overriding

class Organism:
    def sound(self):
        pass

class Human(Organism):
    def sound(self):
        print("Humans communicate via speech, writing,codes, non verbally.")

class Animal(Organism):
    def sound(self):
        print("Animals communicate verbally.")

o = Organism()
o.sound() #not defined / kept abstract 

h = Human()
h.sound()

a = Animal()
a.sound()
```

```python
class Parent:
    def show(self, val):
        print("Parent class show")

class Child(Parent):   # inherit from Parent
    def show(self, val):
        if val == 1:   # use parameter directly
            super().show(val)   # call parent method
        else:
            print("Child class show")

c1 = Child()
c1.show(0)
c1.show(1)
```

```python
#Now parent calls the method of the child

class Parent:
    def show(self, val):
        print("Parent class show")

    def caller(self):
        return "Parent caller."


class Child(Parent):   # inherit from Parent
    def show(self, val):
        if val == 1:   # use parameter directly
            super().show(val)   # call parent method
        else:
            print("Child class show")

    def caller(self):
        return "Child caller."

c1 = Child()
p1 = Parent()

print(p1.caller())
print(c1.caller())

p1 = Child() #parent got reference of Child
print(p1.caller())#parent called the child class caller
 
```

```python
#operator overloading is done use dundar methods

class Points:
    def __init__(self,x):
        self.x = x
#operation over the objects of the same class
    def __add__(self,other):
        return self.x + other.x
    
point1 = Points(10)
point2 = Points(20)
print(point1+point2)
```

```python
# Has a relation

class Engine:
    def start(self):
        print("Engine was started")

class Car:
    def start(self):
        self.engine = Engine() # has a relation
        self.engine.start() #engine starts
        print("car was started.") #then car starts

car = Car()
car.start()
```

```python
#static methods

class Students:
    counter=0
    @staticmethod
    def inc_ctr():
        Students.counter+=1
    @staticmethod
    def show_ctr():
        return Students.counter

    def __init__(self,name,age):
        self.name=name
        self.age=age
        Students.inc_ctr()

aayush = Students("Aayush",22)
gopalJi = Students("GopalJi",22)

print(Students.show_ctr()) #use the class name to use.
```

```python
#making the abstract classes

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def area(self):
        return 4*4
    
sq = Square()
sq.area()
```

```python
# is instance or is subclass of

print(isinstance(sq,Square))
print(issubclass(Square,Shape))
```
