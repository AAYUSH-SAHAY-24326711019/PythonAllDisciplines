```python
# code regarding the tuples and sets

a=() #this is an empty tuple
print("type(a)",type(a))
```

```python
#make a single element tuple
a=(10,)
print(a)
print(type(a))
```

```python
#make the tuple from the list
list_demo = ['a','e','i','o','u']
tuple_demo = tuple(list_demo)
print(tuple_demo, type(tuple_demo))
```

```python
#access the first and last element from the tuple_demo
print(f"first element : {tuple_demo[0]}\nlast element :{tuple_demo[-1]}")
```

```python
#print the mid 3 elements from tuple_demo
print(tuple_demo[1:4])
```

```python
#to iterate through the elements of the tuple
for item in tuple_demo:
    print(item,end=" ")
```

```python
#concatenate 2 tuples
a=(1,2)
b=(3,4)
print(f"concatenation is :{(a+b)}")

#repeat a 2 times
print(f"\nrepeated a 2 times :{(a*2)}")
```

```python
#membership in the tuple
#does 4 belong to tuple
#does 2 belong to tuple

a = (a+b) #done the concatenation
print(a)
print(4 in a)
print(2 in a)
```

```python
# built in functions of tuple
t = tuple([1,2,3,4,5])
print("len(t) :",len(t))
print("sum(t) :",sum(t))
print("min(t) :",min(t))
print("max(t) :",max(t))
```

```python
#to unpack a tuple
a,b,c,d,e = t
print(a ,end=" ")
print(b ,end=" ")
print(c ,end=" ")
print(d ,end=" ")
print(e ,end=" ")
```

```python
#named tuple
from collections import namedtuple
Point = namedtuple("Point",["x","y"])
p = Point(10,20)
print(p.x, p.y)
```

```python
# Sets in python

#They are unordered, no indexing, no duplicate values, they are mutable
#there is a very fast membership testing and supports many mathematical
# functions

#create a set
a= {}  #this is an empty set
a=set(a) #without this line , a is understood to be a dictionary
print(a,type(a))
```

```python
#make the set from the list
list_demo = [10,20,30,40,50]
set_demo = set(list_demo)
print(set_demo) #orientation is managed by internal mechanism
```

```python
#clear the set
set_demo = set()
print(set_demo)
```

```python
set_demo.add(10)
set_demo.add(20)
set_demo.add(30)
set_demo.add(40)
set_demo.add(50)
print(set_demo)
set_demo.pop() #will pop any random element
print(set_demo)
```

```python
#remove 50 
set_demo.discard(50) #you can use remove(value) but error if absent
print(set_demo)
```

```python
# Set operations{ union,intersection,difference,symmetric difference}
a={1,2,3}
b={3,4,5}
print("a|b (a union b) :",(a|b))
print("a&b (a intersection b) :",(a&b))
print("a-b (a difference b) :",(a-b))
print("a^b (a symmetric-difference b) :",(a^b))
```

```python
#all built in methods of sets

#adding elements in set a 
a={10,20,30}
a.add(40)
a.update([50,60,70]) #passing a list of values
print(a)
```

```python
#removing the elements
a.discard(70) #no error if absent
a.remove(60) #error if the value absent
a.pop() #removes a random value
print("after above operations : ",a)
a.clear() #removes all the values
print("after a.clear()",a)
```

```python
# | is union()
# & is intersection()
# - is difference()
# ^ is symmetric_difference()
```

```python
a = {10,20,30,40,50,60,70}
demo = {20,50,60}

print(a.issubset(demo))
print(demo.issubset(a))

print(a.issuperset(demo))
print(a.isdisjoint(demo))

a={10}
demo={30}
print(a.isdisjoint(demo))
```

```python
#test of the membership
print(30 in demo)
```

```python
#make the set containing all the squares from 1 to 10

sq_set = {x*x for x in range(11)}
sq_tuple = tuple(sq_set)
sq_list = list(sq_tuple)
sq_list.sort()
print(sq_list)
```

```python
#make the set containing all the even numbers from 1 to 30

even_set = {x for x in range(31) if(x%2==0)}
even_tuple = tuple(even_set)
even_list = list(even_tuple)
even_list.sort()
print(even_list)
```
