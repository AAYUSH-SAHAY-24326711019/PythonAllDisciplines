```python
# This is the module of Frozen sets and Dictionary

# features of frozenset
# they are immutable, unordered, hashable,support only
# the read only set operations

a = frozenset([1,2,3])
b= frozenset([4,5,6])

# Read only operations include:
print("a|b (a union b) :",(a|b))
print("a&b (a intersection b) :",(a&b))
print("a-b (a difference b) :",(a-b))
print("a^b (a symmetric-difference b) :",(a^b))
```

```python
# Creation of Dictionary

#make an empty dictionary
a={}

#make a dictionary of customer and mob.

contacts = {"Alex":1234,
            "Aayush":2345,
            "GopalJi":3456}

print(contacts)
```

```python
# from a tuple make the dictionary

t = (('x',10),('y',20))

coordinates = dict(t)
print(coordinates)
```

```python
#add key to the coordinates
coordinates['z']=30
print(coordinates)
```

```python
#add multiple keys and values
contacts.update({"person1":11001,"person2":21102})
print(contacts,end="\n")
                
```

```python
# remove the key alex from the contacts
contacts.pop('Alex')
print(contacts)
```

```python
# del will remove a key (old way)
# popitem will remove the last inserted key

contacts.popitem()
print(contacts) # person 2 is removed
```

```python
#get all the keys from the contacts

print(contacts.keys())

#get all the values from the contacts
print(contacts.values())

#get the key value pairs
print(contacts.items())
```

```python
# from the contacts, loop and find the name and the 
# contact with a serial number
i =1
for key,value in contacts.items():
    print(f"sno. {i} Name: {key}  contact: {value}")
    i+=1
```

```python
#using default value

data ={}
data.setdefault("sno",1)

print(data)
```

```python
#make the dictionary of the squares from
#1 to 10

squares = {s:s*s for s in range(11)}
for key,value in squares.items():
    print(f"({key})^2=>{value}")
```

```python
# remove the all items from the dictionary
squares.clear()
print(squares)
```
