This is the lesson for the loops in python

```python
#for loop, printing the values on single line
#default prints in the next line

for i in range (1,6):
    print(i,end=" ")
```

```python
#while loop

i=1
while i<=5:
    print(i,end=" ")
    i+=1

print("The control exits the body of while loop")
```

```python
# do while does not exists
# we can logically derive it

i=1
while True:
    print(i, end=" ")
    i+=1
    if i>6:
        break
```

```python
# making the for each loop

strr = "Aayush"

for s in strr:
    print(s,end="_")
```
