```python
#to add the data to the end of the file
import os
with open(r"temp\new_text.txt","a",encoding="utf-8") as f:
    f.write("new line is added")
```

```python
#to write at the start of the line
import os
with open(r"temp\new_text.txt","r+",encoding="utf-8") as f:
    old = f.read()
    f.seek(0)
    f.write("This is the starting line\n"+old)
    #this is the way , old content is shifted at end
```

```python
#create the csv file and write the data into it
import csv

with open(r"temp\data_records_new.csv","w",newline="")as f:
    writer =csv.writer(f)
    writer.writerow(["name","age","city"])
    writer.writerow(["Aayush-Sahay","22","patna"])
    writer.writerow(["Gopal-Ji","22","patna"])
```

```python
#to read from the csv file
import csv
with open(r"temp\data_records_new.csv","r")as f:
    reader= csv.reader(f)
    for row in reader:
        print(row)
```

```python
#create a tsv file
import csv

with open(r"temp\tsv_records_new.tsv","w",newline="")as f:
    writer =csv.writer(f,delimiter="\t")
    writer.writerow(["name","age","city"])
    writer.writerow(["Aayush-Sahay","22","patna"])
    writer.writerow(["Gopal-Ji","22","patna"])
```

```python
#to read from the tsv file
import csv
with open(r"temp\tsv_records_new.tsv","r")as f:
    reader= csv.reader(f,delimiter="\t")
    for row in reader:
        print(row)
```
