Strings and Substrings in python

```python
#substring retrival 
s="Aayush Sahay"
print(s[0:6])
```

```python
#string methods to find

s="Aayush Sahay"
print(s[:s.find(" ")]) 
#from 0 to 1st space , means fetch me 1st word
```

```python
#find me the second word
s= "Aayush Sahay"
print(s[s.find(" ")+1:])
```

```python
# grep in python , find the occurances

text = "Python is powerful"
print('''"Python" in text :''',"Python" in text)
#if yes then where
print('''text.find("Python"): ''',text.find("Python")+1)

#if found then how many times
print('''text.count("Python"):''',text.count("Python"))

#why to make trouble , use logic
print('''text.startswith("Py"):''',text.startswith("Py"))

print('''text.endswith("ful"):''',text.endswith("ful"))
```

```python
#Regular expressions in python
#we will import 

import re
```

```python
'''
----------Rule 1 -----------
. is anything / any char except newline

\d is for digits from 0 to 9
\D means not a digit

\w is a word character from
a-z , AtoZ, 0-9, _(underscore)

\W means not a word

\s means a whitespace
\S means not a whitespace
------------------------

---------Rule 2 -------
* is for 0 or More (start from 0)
+ is for 1 or more (start from 1)
? is for 0 or 1 (nothing except)
{n} means exactly n
{n,} means n or more
{n,m} means n and m
------------------------

-----Rule 3 --------
^ means the start of the string
$ means the end of the string
\b means word boundart
---------------------

-------Rule 4 ---------
[abc] any thing a or b or c
[a-z] any lowercase character
[^abc] not a or not b or not c
--------------------------

-------Rule 5------------
(abc) capturing group
(?:abc) Non capturing group
'''
```

```python
import re

text="Mob : 1234512345"

result=re.findall(r"\d+",text)
print(result)
#means from the start there are digits, + signify more than one
```
