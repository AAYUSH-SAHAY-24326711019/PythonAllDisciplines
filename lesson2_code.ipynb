this is for the conditionals in python

```python
# using only single if else statement (indentation =4)
a=10
if a%2==0:
    print("number is even")
else:
    print("number is odd")
```

```python
#using the nested if
citizen = True
age =22

if citizen:
    if age>18:
        print(f"The person of {age} can vote, citizen {citizen}")
```

```python
#use of the if elif else
marks =75

if marks>90:
    print("grade A")
elif marks>80:
    print("grade B")
elif marks>70:
    print("grade C")
else:
    print("fail")
```

```python
# demonstration of the select / switch / match case
command = "start"

match command:
    case "start":
        print("The system is started")
    case "pause":
        print("The system is paused.")
    case "terminated":
        print("The system is terminated")
    case _:
        print("invalid command")
```
