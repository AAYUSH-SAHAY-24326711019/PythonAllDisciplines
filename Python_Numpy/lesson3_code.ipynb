```python
#making the array using the random number array

# rand function makes the number between 0 or 1
#  how to apply it

import numpy as np

var = np.random.rand(4) #gives you the random numbers from 0 to 1 between

print(var)
```

```python
#above used the +ve numbers only

var = np.random.randn(4)
print(var)  #gives you the -ve values as well
```

```python
#make use of thr ranf : includes the 0 also

var = np.random.ranf(4)
print(var)
```

```python
# use of the rand int function , will generate the numbers in a given range
# from the minimum and the maximum value

# var  = np.random.randint(min=10,max=20,total=5)
var  = np.random.randint(10,20,5)
print(var)
```

```python
# Data types in numpy array

# many data types stored in a variable
# a numpy array can store many number and string combinations
# it is known as the data type


# bool tru and False
# int , int c , int p , uint unsigned , float ,complex _
# complex 64,128

var = np.array([1,2,3,4])
print("dtype : ",var.dtype)
```

```python
var = np.array([1,2,2.5,3,4])
print("dtype : ",var.dtype)
```

```python
var = np.array(['A','S','D','F'])
print(var.dtype)
```

```python
# conversion of the data types

x = np.array([1,2,3,4])

print("Data Type :",x.dtype)
```

```python
# convert the above into int 8
x = np.array([1,2,3,4],dtype = np.int8)
print("Dtype :",x.dtype)
```

```python
# above is in the int type , it need to be converted into float
x1= np.array([1,2,3,4],dtype="f")
print("Dtype :",x1.dtype)

#the other way

new = np.float32(x1)
print("new dtype is",new.dtype)
#by this you will prevent the changing of the original array

# convert back to int

new = np.int_(new)
print(" again new dtype is",new.dtype)
```

```python
# use of the astype
x3=np.array([1,2,3,4])
new_1 = x3.astype(float)
print(new_1,new_1.dtype)

```
