```python
# Searching operation in the array

import numpy as np

var = np.array([1,2,3,4,2,5,2,5,6,7])

#need to where the 2 , 3 and 4 exists

x = np.where(var == 2)
print("positions of value 2",(x) )

# find the positions of the even values:
x = np.where(var%2==0)
print("positions of even values",(x) )
```

```python
#search the sorted array
#to place the index some value. means when about to place a value at some particular index

#we will use the searchsorted

var = np.array([1,2,3,4,6,7,8,9,10])
#   index       0,1,2,3,4,5,6

x1 = np.searchsorted(var,5)

# if the value 5 is to placed then at what index it will be placed?
# Since 5 is a value that is lying betwen 4 and 6 in above array

print(x1)
print(var)

# This will do the binary search , and suggest you the best
# index number to maintain the search order
```

```python
#now we know that index from beginning can be also converted to the
# index from the end

x1 = np.searchsorted(var,5,side="right")
print(x1)
```

```python
#sorting of the array
var = np.array([2,3,1,5,6,7,0,4,9,8])

print(np.sort(var))
```

```python
vowel = ['e','o','i','a','u']
var = np.array(vowel)

print(np.sort(var))  #this will do the sorting in the ascending order
```

```python
#filtering the array

var3 = np.array(['a','s','d','f'])

filter_value = [True, False, False, True]

new_a = var3[filter_value]

print(new_a) #observation : The values retrieved had their True value inside the filters
```
