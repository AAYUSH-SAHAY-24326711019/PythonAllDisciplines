```python
import numpy as np

ht = np.array([165,166,167,169,170,172])
print("sqrt of ht is = ",np.round(np.sqrt(ht),3)) #make it 3 decimal places
```

```python
var = np.array([45])
print("sin of 45 = ",np.sin(var))
print("cos of 45 = ",np.cos(var))
```

```python
# cumulative sum
var = np.array([1,2,3])
print(np.cumsum(var)) #keep 1 , then for 2 do (1+2)=3 [place at index of 2] now for 3 do (3+3)=6 (final index)
```

```python
#operation on the matrices require them to be of the same shape.
#But sometimes, we can face that they are of different shapes.
#Still we got a way out for this.

#we will get the value error 
#using broadcasting numpy array

V1 = np.array([[1,2,3]])
print("V1 :",V1)

V2 = np.array([
    [1],
    [2],
    [3]
])
print("V2 :")
print(V2)

print()
print(V1.shape)
print(V2.shape)

print()
print(np.multiply(V1,V2))
```
