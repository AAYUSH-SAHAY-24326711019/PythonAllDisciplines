```python
# The use of the multiplication and the dot product

A = np.matrix([[1,2],[1,2]])
B = np.matrix([[1,2],[1,2]])

print(A*B)
print()
print(A.dot(B))

print()
print(np.multiply(A,B)) #inconsistent results
```

```python
The functions that are applicable to the matrices
using numpy are : 
transpose, swap axes, inverse, power, determinate

```

```python
A = np.matrix([
    [1,2,3],
    [4,5,6]
])

print(np.transpose(A))

print("\nShortcut")
print(A.T)
```

```python
# Swap axes
print(np.swapaxes(A,0,1))
```

```python
# inverse of a matrix
A = np.matrix([
    [1,2],
    [3,4]
])

var3 = np.linalg.inv(A)
print(var3)
```

```python
# Adjoint  of a matrix
A = np.array([
    [1, 2],
    [3, 4]
])

det = np.linalg.det(A) # find the determinant [ensure a square matrix]
inv = np.linalg.inv(A)

adj = det * inv
print(adj)
```
