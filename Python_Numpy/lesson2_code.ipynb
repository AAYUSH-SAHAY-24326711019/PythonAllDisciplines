```python
#making an array full of ones
import numpy as np
ar_one = np.ones(4)
print(ar_one)
```

```python
#making an empty array

ar_em = np.empty(4)
print(ar_em) #it wll take the memory of the recently created array
```

```python
ar_rn = np.arange(4)
print(ar_rn) #made the array with the range
```

```python
#making the identity matrix like array
# all digonals are filled with one
ar_dig = np.eye(3) #made a 3 by 3 identity matrix
print(ar_dig)
```

```python
# make the identity matrix using the dimension 3,5
ar_demo = np.eye(3,5)
print(ar_demo)
```
