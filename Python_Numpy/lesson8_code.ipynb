```python
import numpy as np

var = np.array([1,2,3,4,5])

for i in var:
    print(i,end=" ... ")
```

```python
#iteration inside the 2 d array

var1 = np.array([
    [1,2,3,4],
    [1,2,3,4]
    
    ])

#one way
for i in var1:
    print(i)

print()

#the other way

#this will make the entire array into a single line form
for i in var1:
    for j in i:
        print(j,end=" . ")
```

```python
#iteration inside the 3 d array

var1 = np.array([
    [
    [1,2,34],
    [1,2,34],
    [1,2,34]
    ]
    ,
    [
    [1,2,35],
    [1,2,35],
    [1,2,35]
    ]
])

#one way
for i in var1:
    print(i)

print()

#the other way

#this will make the entire array into a single line form
for i in var1:
    for j in i:
        for k in j:
            print(k, end="_ _")

#this is other way
print()
for i in np.nditer(var1):
    print(i,end=" ")
```

```python
#utility :
# we want that the indexing is also shown

# get the data point with the indexes as well

var1 = np.array(
    [
    [1,2,35],
    [1,2,35],
    [1,2,35]
    ]
)

for i,d in np.ndenumerate(var1):
    print(i,d, end="\n")
```
