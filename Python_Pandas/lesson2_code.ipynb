```python
# multiindex

multi_data = pd.DataFrame({
    'Park':['A','A','B','B'],
    'Name' : ['Anna','Ben','Chris','Daisy'],
    'Lemonades':[2,3,4,5]
}).set_index(['Park','Name'])
```

```python
multi_data
```

```python
# find the rows where the lemonades numbers is greater than 3

multi_data[multi_data['Lemonades']>=3]
```

```python
# using the query where you will mention the name of the colun and give the condition

multi_data.query('Lemonades>3')
```
