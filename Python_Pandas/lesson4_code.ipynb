```python
#Data Cleaning and Preprocessing using Pandas

import pandas as pd


sales = pd.Series([2,None,5])

sales
```

```python
sales.dropna()
```

```python
# fill it with a number like 5
sales.fillna(5)
```

```python
#aggregation :substitute the nan with mean

sales.fillna(sales.mean())  #best work than use the min or max
```

```python
# fill the na with the std

sales.fillna(sales.std())
```

```python
sales.fillna(sales.var())
```

```python
# to removes the duplicated entries

name = ['Aayush','Aayush',"Gopal Ji",]
contact= ['123','123',"555",]

data = pd.DataFrame({'Name':name, 'Contact':contact})

data
```

```python
#to remove the duplicates in above. One entry will be kept and other will be removed

data.drop_duplicates() #kept the one with the smaller id.
```

```python
# Replace the values in the data

# replace 123 with 555

data.replace('123',"555",inplace=True) #made it permanent

data

```

```python
data['Contact'] =data['Contact'].astype(float)

data
```
