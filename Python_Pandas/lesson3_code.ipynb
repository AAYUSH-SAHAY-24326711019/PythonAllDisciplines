```python
# to convert the multi data to the csv

import pandas as pd

multi_data = pd.DataFrame({
    'Park':['A','A','B','B'],
    'Name' : ['Anna','Ben','Chris','Daisy'],
    'Lemonades':[2,3,4,5]
}).set_index(['Park','Name'])
```

```python
multi_data.to_csv('lemonade_sales.csv')
```

```python
# to import the csv file

data1 = pd.read_csv('lemonade_sales.csv')

print(data1)
```
