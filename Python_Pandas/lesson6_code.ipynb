```python
import pandas as pd
```

```python
# concatenation and joining of the datasets

s1 = pd.Series(['A','B','A','B','C'])
s2 = pd.Series([10,20,30,40,50])

data1 = pd.DataFrame({'Name':s1, 'Lemonades':s2})

s1 = pd.Series(['A','B','A','B','C'])
s2 = pd.Series([11,22,33,44,55])

data2 = pd.DataFrame({'Name':s1, 'Lemon':s2})
```

```python
data1
```

```python
data2
```

```python
pd.concat([data1,data2])
```

```python
'''above the common columns are counted just only once
The specific areas are left as NaN due to column name is different
and implies for that different columns the data does not exists
in the respective dataframes. 

No values exist inside the Data1 for Lemon 
No values exist inside the Data2 for Lemonade
hence on concatenation the missing values for each other will be
filled as the NaN

'''
```

```python
data1.append(data2)
```

```python
#early versions of pandas had the function for appending the dataframes object
#gives us the results as same as the concat
```

```python
# merging the data1 and data2

pd.merge(data1,data2,on='Name')
```

```python
#this is kind of a 1NF 
#we cannot see NaN here.
#merge shows you the matching records only.
```

```python
#join on index

# join is performed on the basis of the common attribute

#the join is the sql left join some kind of

data1.join(data2.set_index('Name'), on='Name')
```
