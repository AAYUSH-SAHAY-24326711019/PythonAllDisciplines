```python
import pandas as pd

name = ['Aayush','Aayush',"Gopal Ji",]
contact= ['123','123',"555",]

data = pd.DataFrame({'Name':name, 'Contact':contact})

data
```

```python
import pandas as pd

multi_data = pd.DataFrame({
    'Park':['A','A','B','B'],
    'Name' : ['Anna','Ben','Chris','Daisy'],
    'Lemonades':[2,3,4,5]
}).set_index(['Park','Name'])
```

```python
grouped = multi_data.groupby('Name')
grouped.sum()
```

```python
print(grouped['Lemonades'].mean())
```

```python
print(grouped['Lemonades'].min())
print()
print(grouped['Lemonades'].max())
```

```python
multi_data.pivot_table(values='Lemonades',index = 'Name')
#the problem is that it does the computation of the means before display
```

```python
# crosstab

pd.crosstab(multi_data.index.get_level_values('Name'), multi_data['Lemonades'])

#you are doing the fetch incase the Name is set as an index.
```
